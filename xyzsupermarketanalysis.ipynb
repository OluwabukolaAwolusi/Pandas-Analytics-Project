### Data Visualization

```python
sns.countplot(x='Product line', data=cxyz)
plt.title('Count of Products by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Branch', data=cxyz)
plt.title('Count of Branch by sales')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='City', data=cxyz)
plt.title('Count of City by sales')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

### This shows that lagos has the highest sales

```python
sns.countplot(y='Product line', data=cxyz)
plt.title('Highest and lowest sold product line')
plt.xlabel('Branch')
plt.ylabel('Product line')
plt.show()
```

```python
sns.countplot(x='Payment',hue='Branch', data=cxyz)
plt.title('Payment channel per branch')
plt.xlabel('Payment')
plt.ylabel('Branch')
plt.show()
```

```python
sns.countplot(x='Payment', data=cxyz)
plt.title('Payment channel per branch')
plt.xlabel('Payment')
plt.ylabel('Branch')
plt.show()
```

```python
sns.boxplot(x='Branch', y='Rating', data=cxyz)
plt.title('Rating per Branch')
plt.xlabel('Branch')
plt.ylabel('Rating')
plt.show()
```

```python
sns.catplot(x='Product line', y='Quantity', hue='Gender', data=cxyz, kind='bar',aspect=4)
plt.title('Quantity of Products by Product Line and Gender')
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.show()
```

```python
sns.catplot(x='Product line', y='Total', hue='Gender', data=cxyz, kind='bar',aspect=4)
plt.title('Quantity of Products by Product Line and Gender')
plt.xlabel('Product Line')
plt.ylabel('Total')
plt.show()
```

## summary of the product line by gender
* It was observed that across all cities where the supermarket is located, women purchase;
* more fashion accessories,compared to males, while male have a higher habit of purchasing sports and lesure, and health and beauty women also record a high total purchase of home and life style products
* High quantity of food and beverages a slight difference is seen in the quantity of electronic accessories product between male and female

```python
sns.catplot(x='Product line', y='Unit price', data=cxyz, kind='point',aspect=4)
plt.title('Product line by unit price')
plt.xlabel('Product Line')
plt.ylabel('Unit price')
plt.show()
```
